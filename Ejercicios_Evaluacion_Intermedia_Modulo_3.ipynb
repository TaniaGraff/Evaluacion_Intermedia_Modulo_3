- Usando el dataframe de los albumes que hemos creado en el ejercicio anterior. ¿Cuál
es la media y la desviación estándar de "danceability", "acousticness" y
"speechiness" de cada artista? Muesrtra los resultados en tres dataframes
diferentes, uno para cada métrica ("danceability", "acousticness" y "speechiness").
Ordena los resultados de mayor a menor en base a la media.

```python
#Calculo la media y la desviación estándar para 'danceability'.
danceability = df_album.groupby('artist')['danceability'].agg(['mean', 'std']).sort_values(by='mean', ascending=False).sort_values(by='mean', ascending=False)
danceability
```

```python
#Hago lo mismo para 'acousticness'.
acousticness= df_album.groupby('artist')['acousticness'].agg(['mean', 'std']).sort_values(by='mean', ascending=False)
acousticness
```

```python
#Lo repito para 'speechiness'.
speechiness = df_album.groupby('artist')['speechiness'].agg(['mean', 'std']).sort_values(by='mean', ascending=False)
speechiness
```

```python
#Uno todos los datos en un único DF.
df_media_desviacion = df_album.groupby('artist')[['danceability', 'acousticness', 'speechiness']].agg(['mean', 'std']).reset_index()
df_ordenado = df_media_desviacion.sort_values(by=('danceability', 'mean'), ascending=False)
df_ordenado
```

- Ahora queremos analizar la colaboración entre artistas en la industria musical, se
tienen datos de las colaboraciones realizadas en los últimos años. Se desea conocer
cuáles son los artistas que han colaborado en mayor cantidad de canciones y
generar un ranking con los diez primeros. ¿Qué artistas con los que más
colaboraciones tienen? Para esto tendréis que:
    - Filtra los datos para quedarte solo con aquellos datos donde en la columna
"track" tengamos "feat".
    - Cuenta cuantas canciones tiene cada artista utilizando el dataframe generado
en el paso anterior.
    - Debes contestar a las siguientes preguntas:
    - ¿Cuáles son los diez artistas que más han colaborado según los datos
que tenemos?
    - ¿Cuál es el número de canciones en los que ha colaborado el artista que
ocupa el primer lugar en el ranking?
    - ¿Cuál es el artista que ocupa el décimo lugar en el ranking y en cuántas
canciones ha colaborado?

```python
#Busco las canciones que en el título contengan la palabra 'Feat'.
df_colaboraciones = df_spoty[df_spoty['track'].str.contains('Feat')]
df_colaboraciones.head()
```

```python
#Ordeno a los artistas en función del númeor de colaboraciones que han hecho.
canciones_por_artista = df_colaboraciones.groupby('artist')['track'].count().reset_index(name='number_colaborations')
canciones_por_artista
```

```python
canciones_por_artista.sort_values(by='number_colaborations', ascending=False).head(10)
```

```python
#Hago un barplot para mostrar los resultados de forma más visual.
top_ten_colab = canciones_por_artista.sort_values(by='number_colaborations', ascending=False).head(10)


sns.barplot(x= 'artist', y='number_colaborations', data= top_ten_colab, palette= 'magma', order=top_ten_colab.sort_values('number_colaborations', ascending=False)['artist'])

#Pongo título a la gráfica y cambio los nombres de los ejes.
plt.title('Colaboraciones por artista')
plt.xlabel("Artistas")
plt.ylabel("Nº Colaboraciones")

plt.xticks(rotation=90);
```

```python
print(f'El artista con más colaboraciones es:\n {top_ten_colab.max()}')
print('.............................................................')
print(f'El artista con menos colaboraciones es:\n {top_ten_colab.min()}')
```
