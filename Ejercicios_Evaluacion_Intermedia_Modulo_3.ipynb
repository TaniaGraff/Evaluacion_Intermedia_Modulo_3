```python
#Compruebo que la columna 'track' es la que contiene el t√≠tulo de las canciones.
df_spoty['track']

#Creo una funci√≥n para unificar el t√≠tulo de las canciones.
def unificar_titulos(df, columna):
    df[columna] = df[columna].str.title()

#Llamo a la funci√≥n.
unificar_titulos(df_spoty, 'track')

#Compruebo que el DF se ha modificado correctamente.
df_spoty['track']
```

```python
#Compruebo la columna de 'artist' buscando valores an√≥malos y no encuentro ninguno. 
list(df_spoty['artist'].unique())
```

```python
#A√∫n as√≠, realizo la funci√≥n que me piden.
def limpiar_nombre(df, columna):
    df[columna] = df[columna].str.replace('\$', 's', regex=True)

#Llamo a la funci√≥n.
limpiar_nombre(df_spoty, 'artist')

#Y vuelvo a comprobar que los datos son correctos.
list(df_spoty['artist'])
```

- Ahora queremos analizar la colaboraci√≥n entre artistas en la industria musical, se
tienen datos de las colaboraciones realizadas en los √∫ltimos a√±os. Se desea conocer
cu√°les son los artistas que han colaborado en mayor cantidad de canciones y
generar un ranking con los diez primeros. ¬øQu√© artistas con los que m√°s
colaboraciones tienen? Para esto tendr√©is que:
    - Filtra los datos para quedarte solo con aquellos datos donde en la columna
"track" tengamos "feat".
    - Cuenta cuantas canciones tiene cada artista utilizando el dataframe generado
en el paso anterior.
    - Debes contestar a las siguientes preguntas:
    - ¬øCu√°les son los diez artistas que m√°s han colaborado seg√∫n los datos
que tenemos?
    - ¬øCu√°l es el n√∫mero de canciones en los que ha colaborado el artista que
ocupa el primer lugar en el ranking?
    - ¬øCu√°l es el artista que ocupa el d√©cimo lugar en el ranking y en cu√°ntas
canciones ha colaborado?

```python
#Busco las canciones que en el t√≠tulo contengan la palabra 'Feat'.
df_colaboraciones = df_spoty[df_spoty['track'].str.contains('Feat')]
df_colaboraciones.head()
```

```python
#Ordeno a los artistas en funci√≥n del n√∫meor de colaboraciones que han hecho.
canciones_por_artista = df_colaboraciones.groupby('artist')['track'].count().reset_index(name='number_colaborations')
canciones_por_artista
```

```python
canciones_por_artista.sort_values(by='number_colaborations', ascending=False).head(10)
```

```python
#Hago un barplot para mostrar los resultados de forma m√°s visual.
top_ten_colab = canciones_por_artista.sort_values(by='number_colaborations', ascending=False).head(10)


sns.barplot(x= 'artist', y='number_colaborations', data= top_ten_colab, palette= 'magma', order=top_ten_colab.sort_values('number_colaborations', ascending=False)['artist'])

#Pongo t√≠tulo a la gr√°fica y cambio los nombres de los ejes.
plt.title('Colaboraciones por artista')
plt.xlabel("Artistas")
plt.ylabel("N¬∫ Colaboraciones")

plt.xticks(rotation=90);
```

```python
print(f'El artista con m√°s colaboraciones es:\n {top_ten_colab.max()}')
print('.............................................................')
print(f'El artista con menos colaboraciones es:\n {top_ten_colab.min()}')
```
